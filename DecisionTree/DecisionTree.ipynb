```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
df = pd.read_csv('Decision-Tree-Classification-Data.csv')
df
```

```python
X = df.drop('diabetes', axis = 1)
Y = df.diabetes
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.28, random_state=43)
```

```python
Decision_Tree_Class_Model = DecisionTreeClassifier()
Decision_Tree_Class_Model.fit(X_train, Y_train)
```

```python
Y_pred = Decision_Tree_Class_Model.predict(X_test)
Y_pred
```

```python
accuracy = metrics.accuracy_score(Y_test, Y_pred)
print("Accuracy : ", accuracy)
```

```python
Decision_Tree_Class_Model = DecisionTreeClassifier(criterion="entropy",max_depth=7)
Decision_Tree_Class_Model.fit(X_train, Y_train)
Y_pred = Decision_Tree_Class_Model.predict(X_test)
print("Accuracy : ", metrics.accuracy_score(Y_test, Y_pred))
```

```python
ddf = pd.read_csv('diabetes.csv')
ddf
```

```python
ddf.info()
```

```python
feature_cols = ['Pregnancies','Glucose','BloodPressure','Insulin','BMI','DiabetesPedigreeFunction','Age']
x = ddf[feature_cols]  # Features
y = ddf.Outcome   # Target variable
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size= 0.2, random_state=30)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf.fit(x_train,y_train)
```

```python
#Predict the response for test dataset
y_pred = clf.predict(x_test)
y_pred
```

```python
accuracy = metrics.accuracy_score(y_test,y_pred)
print("Accuracy : ", accuracy)
```

```python
clf1 = DecisionTreeClassifier(criterion="entropy", max_depth=3)
clf1.fit(x_train,y_train)
y_pred1 = clf1.predict(x_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred1))
```
